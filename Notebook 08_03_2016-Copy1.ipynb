```python
df1 = df[df["Total Assets"]>0]
df1
```

```python
df1["Bank Name"].nunique()
```

```python
df1["Net Charge-off"] = df1["Total Loan Charge-offs"] - df1["Total Loan Recoveries"]

df1["ROA"] =df1["Net Income"]/df1["Total Assets"]
df1["ROE"] = df1["Net Income"]/df1["Total Equity"]
df1["Loan-to-Assets"] = df1["Total Loan, net of unearned income"]/df1["Total Assets"]
df1["Fixed Assets-to-Assets"] = df1["Fixed Assets"]/df1["Total Assets"]
df1["Net Charge Off-to-Loans"] = df1["Net Charge-off"]/df1["Total Loan, net of unearned income"]
df1["Bank Size"] = np.log(df1["Total Assets"].astype("float64"))
df1["Liquidity"] = (df1["Cash"]+df1["Available-for-sale securities"])/df1["Total Assets"]
df1["Loan Loss Provision %"] = df1["Loan Loss Provision"]/df1["Total Loan, net of unearned income"]
```
